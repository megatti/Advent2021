```python
def snailaccess(snail_number, coordinate):
    """Goes to coordinate of snail_number?"""
    eval_string = "snail_number"
    for i in coordinate:
        eval_string += f"[{i}]"
    # Now I have something like "snail_number[0][1][0]"
    # Return this value, or return this string?
    return eval_string

def snailset(snail_number, coordinate, value):
    """Sets the element at coordinate of snail_number to value."""
    eval_string = snailaccess(snail_number, coordinate)
    eval_string += f" = {value}"
    exec(eval_string)
    return snail_number

def snailget(snail_number, coordinate):
    """Gets the value at coordinate of snail_number."""
    eval_string = snailaccess(snail_number, coordinate)
    value = eval(eval_string)
    return value
```

```python
def find_left(snail_number, coordinate):
    """Works from snail number and the coordinate of some number to find the left regular number.
    Coordinate will be a list of 1s and 0s, each meaning the index.
    Returns the coordinate of the left number."""
    if 1 not in coordinate:
        # Already on the leftmost element, so no left-element :/
        return None
    else:
        last_right = len(coordinate) - (list(reversed(coordinate)).index(1) + 1)  # the last time the snail_number went to right
        change_coord = coordinate[:last_right]
        # need to traverse to the left once, then right
        change_coord.append(0)
        while True:
            try:
                # try getting the right number
                snailget(snail_number, change_coord + [1])
                # if it returns something, then there is either a value or a list here
                # keep going
                change_coord.append(1)
            except TypeError:
                # Couldn't get right value since value at change_coord is just a number
                # Return the coordinate of this number
                return change_coord
            

def find_right(snail_number, coordinate):
    """Works from snail number and the coordinate of some number to find the right regular number.
    Coordinate will be a list of 1s and 0s, each meaning the index.
    """
    if 0 not in coordinate:
        # Already on the rightmost element, so no right-element :/
        pass
    else:
        last_left = len(coordinate) - (list(reversed(coordinate)).index(0) + 1)
        change_coord = coordinate[:last_left]
        # need to traverse to the right once, then left
        change_coord.append(1)
        while True:
            try:
                # try getting the left number
                snailget(snail_number, change_coord + [0])
                # if it returns something, then there is either a value or a list here
                # keep going
                change_coord.append(0)
            except TypeError:
                # Couldn't get right value since value at change_coord is just a number
                # Return the coordinate of this number
                return change_coord
    pass
```

```python
def snailexplode(snail_number, coordinate):
    """Explodes the pair at some coordinate. Only works if pair to explode is just a pair of numbers."""
    explode_pair = snailget(snail_number, coordinate)
    if not all(isinstance(i, int) for i in explode_pair):
        # One of this pair isn't a list
        raise TypeError(f"One of the pair at this coordinate is a list, not an integer. Coord: {coordinate}")
    else:
        # Do the exploding
        left_coord = find_left(snail_number, coordinate)
        if left_coord is None:
            # no left value
            pass
        else:
            left_value = snailget(snail_number, left_coord)
            snail_number = snailset(snail_number, left_coord, left_value + explode_pair[0])
            
        # Do the exploding
        right_coord = find_right(snail_number, coordinate)
        if right_coord is None:
            # no left value
            pass
        else:
            right_value = snailget(snail_number, right_coord)
            snail_number = snailset(snail_number, right_coord, right_value + explode_pair[1])
            
    snail_number = snailset(snail_number, coordinate, 0)
    return snail_number


def snailsplit(snail_number, coordinate):
    """Splits the integer at some coordinate. Only works if value at coordinate is an integer."""
    split_value = snailget(snail_number, coordinate)
    if not isinstance(split_value, int):
        # Value isn't an integer
        raise TypeError(f"Value to split at this coordinate is a list, not an integer. Coord: {coordinate}")
    else:
        floor_half = int(math.floor(split_value/2))
        ceil_half = int(math.ceil(split_value/2))
        snail_number = snailset(snail_number, coordinate, [floor_half, ceil_half])
        
    return snail_number
```

```python
# make two functions, one to check for explosions, the other to check for splits

def check_explode(orig_number, snail_number, coordinate):
    # orig number is the original thing, snail number is a subsection of it for recursion
    # print(snail_number)
    if all(isinstance(i, int) for i in snail_number):  # if snail_number is just a pair of numbers
        # print("pair of ints found", snail_number)
        if len(coordinate) >= 4: # if more than 4 deep
            # this coordinate needs to be passed to the exploding function
            # or maybe just explode it now and then return back since it needs to start again?
            orig_number = snailexplode(orig_number, coordinate)
            return [True, orig_number]
            
    try:
        exploded = check_explode(orig_number, snail_number[0], coordinate + [0])
        if exploded[0]:
            return exploded
    except TypeError:
        pass
        # print(snail_number[0])
        # first element is just a number
    
    try:
        exploded = check_explode(orig_number, snail_number[1], coordinate + [1])
        if exploded[0]:
            return exploded
    except TypeError:
        pass
        # print(snail_number[1])
        
    # went through both and hasn't exploded anything
    return [False, orig_number]

explode_test = [[[[[9,8],1],2],3],4]
data = [True, []]
while data[0]:
    data = check_explode(explode_test, explode_test, [])
print(data)
print()

explode_test = [[3,[2,[1,[7,3]]]],[6,[5,[4,[3,2]]]]]
data = [True, []]
while data[0]:
    data = check_explode(explode_test, explode_test, [])
print(data)
print()

explode_test = [[[[9,8],1],2],3]
data = [True, []]
while data[0]:
    data = check_explode(explode_test, explode_test, [])
print(data)
print()
```

```python
def check_split(orig_number, snail_number, coordinate):
    # orig number is the original thing, snail number is a subsection of it for recursion
    # print(snail_number)
    if isinstance(snail_number, int):  # if snail_number is just a number
        if snail_number >= 10:
            orig_number = snailsplit(orig_number, coordinate)
            return [True, orig_number]
            
    try:
        splitted = check_split(orig_number, snail_number[0], coordinate + [0])
        if splitted[0]:
            return splitted
    except TypeError:
        pass
        # print(snail_number[0])
    
    try:
        splitted = check_split(orig_number, snail_number[1], coordinate + [1])
        if splitted[0]:
            return splitted
    except TypeError:
        pass
        # print(snail_number[1])
        
    # went through both and hasn't split anything
    return [False, orig_number]

split_test = [[[[[9,8],1],2],3],4]
data = [True, []]
while data[0]:
    data = check_split(split_test, split_test, [])
print(data)
print()

split_test = [[[[[10,8],1],2],3],4]
data = [True, []]
while data[0]:
    data = check_split(split_test, split_test, [])
print(data)
a, b = data
print(a)
print()
```

```python
# I now have two functions, one for checking for pairs needing exploding,
# one for checking for values that need splitting

def snailsum(pair1, pair2):
    """Adds two pairs together. This will involve first just making the combined list, then resolving that."""
    summed = [pair1, pair2]
    result = snailresolve(summed)
    return result

def snailresolve(snail_number):
    # Go through snail_number
    # 1: Explode leftmost pair nested inside four pairs
    # 2: Split numbers greater than 10
    # Do this until there are no pairs that fit these criteria left
    
    while True:
        # First, go through and explode all pairs that need it until there are none left?
        data = [True, []]
        while data[0]:
            data = check_explode(snail_number, snail_number, [])
        snail_number = data[1]
        
        # Then, go through and split a number if needed
        data = [False, []]
        splitted, snail_number = check_split(snail_number, snail_number, [])
        # If a number was splitted, go back to checking for explosions
        if splitted:
            continue
        # otherwise, no number needed splitting, so this number has been resolved
        break
    # snail number has been resolved, so just return it
    return snail_number

result = lines[0]
for index, pair in enumerate(lines[1:], 1):
    result = snailsum(result, pair)
    print(result)
```

```python
def snailmagnitude(snail_number):
    if isinstance(snail_number, int):
        # snail_number is just an integer
        return snail_number
    else:
        # snail_number is a list
        left_mag = 3 * snailmagnitude(snail_number[0])
        right_mag = 2 * snailmagnitude(snail_number[1])
        return left_mag + right_mag
        
print(snailmagnitude([9,1]))
print(snailmagnitude([1,9]))
print(snailmagnitude([[9,1],[1,9]]))
```
