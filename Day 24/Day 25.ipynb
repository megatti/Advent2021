```python
def parse_step(variables, step, monad_input, input_count):
    if step[0] == "inp":
        variables[step[1]] = int(monad_input[input_count])
        input_count += 1
    elif step[0] == "add":
        try:
            variables[step[1]] += variables[step[2]]
        except KeyError:
            variables[step[1]] += step[2]
    elif step[0] == "mul":
        try:
            variables[step[1]] *= variables[step[2]]
        except KeyError:
            variables[step[1]] *= step[2]
    elif step[0] == "div":
        try:
            variables[step[1]] = variables[step[1]] // variables[step[2]]
        except KeyError:
            variables[step[1]] = variables[step[1]] // step[2]
    elif step[0] == "mod":
        try:
            variables[step[1]] = variables[step[1]] % variables[step[2]]
        except KeyError:
            variables[step[1]] = variables[step[1]] % step[2]
    elif step[0] == "eql":
        try:
            variables[step[1]] = int(variables[step[1]] == variables[step[2]])
        except KeyError:
            variables[step[1]] = int(variables[step[1]] == step[2])
    return variables, input_count


def gen_input():
    return [random.randint(1, 9) for _ in range(14)]
```
