```python
import collections

pairs = collections.defaultdict(str)

with open("polymer.txt") as f:
    lines = [line.strip() for line in f.readlines()]
    base = lines[0]
    for line in lines[2:]:
        start = line.split(" -> ")[0]
        end = line.split(" -> ")[1]
        pairs[start] = end
        
print(base)
print(pairs)
```

```python
def step_through(base, pairs):
    parts = [i + pairs[i + j] for i, j in zip(base,base[1:])]
    parts.append(base[-1])
    return "".join(parts)
```

```python
old_base = base

for i in range(10):
    base = step_through(base, pairs)

counts = collections.defaultdict(int)
for char in "".join(set(base)):
    counts[char] = base.count(char)
    
# max_count = max(counts, key=counts.get)
# min_count = min(counts, key=counts.get)

print(max(counts.values()) - min(counts.values()))

base = old_base
```

Ok. I have a dictionary that shows how each base pair will change after 20 steps. If I find the base after 20 steps, and use the dictionary to see how the count of each letter will change after 20 steps, maybe I can just see how the count changes 20 steps after 20 steps, i.e. after 40 steps? that sounds reasonable to me :o

```python
print(step_20.keys())

count_after_20 = {}

for key in step_20.keys():
    print(key)
    value = step_20[key][:-1]
    # take off the last letter because that'll be the first letter in the next pair, so it'll get counted next time
    
    # Count how many occurences of each character after 20 steps
    # First, let's just see if it can do this without running out of memory lol ;_;
    counts = collections.defaultdict(int)
    for char in "".join(set(value)):
        counts[char] = value.count(char)
        
    print(counts)
    
    count_after_20[key] = counts
```

```python
print(count_after_20)

old_base = base

for i in range(2):
    print(i)
    base_10_steps = []

    for i, j in zip(base, base[1:]):
        base_10_steps.append(step_10[i + j][:-1])
    
    base_10_steps.append(base[-1])
    base = "".join(base_10_steps)
    
base_20 = base
base = old_base

# now I have the chain after 20 steps
# go through each pair

count_each_letter = collections.defaultdict(int)

for i, j in zip(base_20, base_20[1:]):
    pair = i + j
    for key, value in count_after_20[pair].items():
        count_each_letter[key] += value
        
print(count_each_letter)
```
